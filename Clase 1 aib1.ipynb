```python
#condicionaless con if elif y else
edad = int(input('Dime tu edad'))
if(edad<18):
    print('Menor de Edad')
elif(edad>= 18 and edad<65):
    print("En edad de trabajar")
elif(edad>=45 and edad<65):
    print('Edad media')
else:
    print('Esta jubilado')
    print('Por fin')
```

```python
#condicionales anidados
#condicionaless con if elif y else
edad = int(input('Dime tu edad'))
if(edad<18):
    print('Menor de Edad')
elif(edad>= 18 and edad<65):
    print("En edad de trabajar")
    if(edad>=18 and edad<45):
        print('Edad media 1')
elif(edad>=45 and edad<65):
    print('Edad media')
else:
    print('Esta jubilado')
    print('Por fin')
```
